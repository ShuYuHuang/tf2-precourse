```python
import tensorflow as tf
import numpy as np
```

## Math
基本數學運算

```python
A = np.random.random((3, 4))
B = 3
A
```

```python
tf.add(A, B)
```

```python
tf.subtract(A, B)
```

```python
tf.multiply(A, B)
```

```python
tf.divide(A, B)
```

```python
tf.exp(A)
```

```python
tf.math.log(A)
```

```python
tf.greater(A, B)
```

```python
tf.less(A, B)
```

```python
tf.equal(A, B)
```

## Matrix Operation
矩陣的運算，矩陣轉置、矩陣乘法

```python
A = np.random.random((3, 4))
B = np.random.random((4, 1))
A
```

```python
tf.transpose(A)
```

```python
tf.matmul(A, B)
```

## Tensor Operations
張量的操作，這邊指的是可包含三個軸以上的向量

```python
A = np.random.random((3, 4))
A
```

### Casting
可以用tf.cast將資料轉換型別

```python
print('original dtype of A:', A.dtype)
tf.cast(A, 'float32')
```

### 形狀操作
可以用tf.reshape來改變資料形狀，-1代表依照剩餘量計算，只能放在一個軸

```python
tf.reshape(A, (-1, 1))
```

可以用tf.expand_dims來增加一個冗餘軸，在軸有指定需求時可以拿來調整

```python
tf.expand_dims(A, -1)
```

可以用tf.squeeze來減少冗餘的軸，在軸有指定需求時可以拿來調整

```python
tf.squeeze(tf.expand_dims(A, -1))
```

### Slicing & Joining
在資料要從不同來源合併或分割時可以用到

```python
A = np.random.random((3, 4))
B = 3
A
```

tf.slice可從Tensor中開頭為begin 的地方，切出大小為size的內容!

```python
tf.slice(A, begin=[1, 1], size=[2, 3])
```

image.png

tf.split可以將一個Tensor切為等份的數個Tensor

```python
tf.split(A, 3, axis=0)
```

```python
tf.split(A, 2, axis=1)
```

tf.concat可以將兩個以上Tensor延著某個軸接起來

```python
tf.concat((A, A), axis=0)
```

```python
tf.concat((A, A), axis=1)
```

tf.tile可以將Tensor依指定的軸重複指定次數

```python
tf.tile(A, [2, 3])
```

image.png

tf.stack可以將兩個以上Tensor延著一個新的軸接起來

```python
tf.stack((A, A), axis=2)
```
